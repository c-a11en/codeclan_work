````python
# module imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# data read
demographic_data = pd.read_csv("data/life_expectancy_and_income.csv")
````

````python
demo_data_2019 = (
    demographic_data
    .loc[demographic_data.year == 2019]
    .copy()
)
````

````python
demo_data_2019
````

For the year 2019, how did life expectancy compare to income for each country?

````python
# base plot
plt.scatter(
    x='income_per_person',
    y='life_expectancy',
    data=demo_data_2019
)

# axis labels
plt.xlabel('Income per person ($)')
plt.ylabel('Life Expectancy (years)')

# title
plt.title('Life Expectancy vs Income in 2019')

# set limits (this plot's probably fine but just to practice)
plt.xlim(0, 80000)
plt.ylim(0, 100)

# control tick marks (original plot was fine but if you wanted more control)
plt.yticks(list(range(0,110,10)))

# just shows plot, no rendering code
plt.show()
````

Histograms. Same plotting options. Different plt._plot_type_()

What does the distribution of income per person look like for 2019?

````python
plt.hist(
    x='income_per_person',
    data=demo_data_2019
)

plt.xlabel('Income per person ($)')
plt.ylabel('Count')
plt.title('Distribution of Income per person for 2019')

plt.show()
````

````python
plt.boxplot(
    x='income_per_person',
    data=demo_data_2019,
    vert=False
)

plt.xlabel('Income per person ($)')
plt.title('Distribution of Income per person for 2019')

plt.show()
````

## Seaborn

For the year 2019, how did life expectancy compare to income for each country?

````python
sns.scatterplot(
    x='income_per_person',
    y='life_expectancy', 
    data=demo_data_2019
)

plt.xlabel('Income per person ($)')
plt.ylabel('Life Expectancy (years)')

plt.show()
````

Seaborn also has functions for other types of plot

- histograms
- bar charts
- boxplots

````python
sns.histplot(x='income_per_person', data=demo_data_2019)

plt.show()
````

````python
sns.boxplot(x='income_per_person', data=demo_data_2019)

plt.show()
````

What was the income per person for the g7 nations in 2019?

````python
g7 = ['Canada', 'Italy', 'France', 'Germany',
      'United Kingdom', 'United States', 'Japan']
````

````python
g7_2019 = (
    demo_data_2019
    .loc[demo_data_2019.country.isin(g7)]
    .sort_values('income_per_person')
    .copy()
)

g7_2019
````

````python
# default
sns.barplot(
    x='country',
    y='income_per_person',
    data=g7_2019
)
````

````python
# flipped axes
sns.barplot(
    y='country',
    x='income_per_person',
    data=g7_2019,
    orient='h'
)
````

````python
# rotating xticks (xlabels)
sns.barplot(
    x='country',
    y='income_per_person',
    data=g7_2019,
    color='steelblue'
)

plt.xticks(rotation=45)
plt.show()
````

### Task
1. Read in the `occupation_prestige` data.
1. Create a plot to show the relationship between `prestige` and `education`.
1. Create a plot to show the minimum `prestige` score for each occupation `type` (i.e. the type variable).
1. Add a title and new axis labels to your 2nd graph.

````python
occupation_prestige = pd.read_csv("data/prestige_occupation.csv")
````

````python
sns.scatterplot(
    x='prestige',
    y='education',
    data=occupation_prestige
)
````

````python
min_prestige = (
    occupation_prestige
    .groupby('type')
    .prestige
    .min()
    .copy()
    .reset_index(name = 'min_prestige') #converts series into dataframe
    .sort_values('min_prestige')
)

min_prestige
````

````python
sns.barplot(
    x='type',
    y='min_prestige',
    data=min_prestige,
    color='steelblue'
)
````

## Colours and Themes (and also other aesthetic properties)

- making colour aesthetics depend on data
- making size, shape aesthetics depend on data
- seaborn themes

````python
sns.scatterplot(
    x='income_per_person',
    y='life_expectancy',
    data=demo_data_2019,
    # hue: colour depends on data
    hue='fertility_rate',
    # set colour palette
    palette='viridis_r',
    # size: size depends on data
    size='fertility_rate'
)
````

Categorical/discrete colours

What if we want to draw attention to countries where the fertility rate is particularly high? 

Here, we'll need to do some pre-plot wrangling (make a high_fertility_rate column)

high_fertility_rate when fertility_rate > 5

````python
import numpy as np
````

```python
np.where(condition, value_if_true, value_if_false)
```

````python
demo_data_2019['high_fertility_rate'] = np.where(
    demo_data_2019.fertility_rate > 5, True, False
)

demo_data_2019.head()
````

````python
sns.scatterplot(
    x='income_per_person',
    y='life_expectancy',
    data=demo_data_2019,
    # hue: colour depends on data
    hue='high_fertility_rate',
    # apply a qualitative / categorical colour palette
    palette='colorblind'
)
````

````python
sns.scatterplot(
    x='income_per_person',
    y='life_expectancy',
    data=demo_data_2019,
    # hue: colour depends on data
    hue='high_fertility_rate',
    # or apply our own
    palette={True: 'tomato', False: 'seagreen'},
    # change the shape of our points
    style="high_fertility_rate"
)
````
