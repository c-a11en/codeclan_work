````python
# practice data
demo_missing = pd.read_csv("data/life_expectancy_and_income_missing.csv")
````

````python
demo_missing.isna().sum()
````

````python
demo_missing.shape
````

````python
# drops all na values from all columns
demo_missing.dropna().shape
````

````python
# drop for a specific column, can persist this with 'inplace=True'
demo_missing.dropna(subset = ["life_expectancy"], inplace=True)
````

````python
demo_missing.isna().sum()
````

````python
# load original data back in
demo_missing = pd.read_csv("data/life_expectancy_and_income_missing.csv")
````

````python
# 3. imputation (simple) - replace NA values with 0
demo_missing.fillna(value = 0)
````

```python
demo_missing.fillna(
    value = {
        'column_with_nas': value_to_replace_na_with
    }
)
```

````python
# 3. imputation (more complex)
demo_missing.fillna(
    value = {
        'life_expectancy': 0,
        'income_per_person': 50
    }
)

demo_missing.fillna(
    value = {
        'life_expectancy': demo_missing.life_expectancy.mean(),
        'income_per_person': demo_missing.income_per_person.median()
    }
)
````

````python
# 3. imputation (complex) replace life expectancy missing values with the 
# mean life expectancy for that country

# 3.1 get mean life expectancy for each country (want this to be the same
# length as our data) - transform method

mean_per_country = demo_missing.groupby('country').life_expectancy.transform('mean')

demo_missing.fillna(
    value = {
        'life_expectancy': mean_per_country,
        'income_per_person': demo_missing.income_per_person.median()
    }
)
````

Tying this all together

````python
(
    occupation_prestige
    .fillna(value = {
        'women': 0
    })
    .loc[(occupation_prestige.type != 'other'), ['job', 'income', 'women', 'type']]
    .assign(income_1000s = occupation_prestige.income / 1000)
    .groupby('type')
    .income_1000s
    .mean()
)
````
